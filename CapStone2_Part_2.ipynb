```python
import pandas as pd
```

First, I load the dataframe conataing the postal codes etc. that created in the Part I notebook. I also load the .csv file containing the cooridnates. 

```python
dfToronto = pd.read_csv('Toronto_PostalCodes.csv')  # dataframe created in Part I notebook

geoData_remote = 'https://cocl.us/Geospatial_data'  # link to coordinates provided in Coursera
geoData = pd.read_csv(geoData_remote)  # load dataframe containing coordinates

# Output column names. It is important to make sure that the column on 
# which the dataframes are later to be merged have the same names
geoData.columns  
```

The dataframe are merged on matching values in the columns PostalCode. I use an inner merge (join) such that the resulting merged dataframe only contains rows corresponding to postal codes that appeared in both dataframes.

```python
# Change column name to align between the two dataframes
geoData.rename(columns={'Postal Code': 'PostalCode'}, inplace=True)

# I use the merge method to merge the two dataframes. 
dfALL = dfToronto.merge(geoData, how='inner', on='PostalCode')
```
